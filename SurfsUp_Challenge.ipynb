```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
# Base.classes.keys()
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
dates = session.query(Measurement.date, Measurement.tobs).all()
for date in dates:
    print(date)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
year = 2010
AllTheJunes = []
while (year <= 2017):
    junes = session.query(Measurement.date, Measurement.tobs).filter(Measurement.date >= f'{year}-06-01' ).filter(Measurement.date <= f'{year}-06-30').all()
    AllTheJunes.append(junes)
    year += 1
print(len(AllTheJunes))
```

```python
#  3. Convert the June temperatures to a flat list.
june_temps = sum(AllTheJunes,[])
print(june_temps)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temps_df = pd.DataFrame(june_temps, columns=['Date', 'June Temp (degF)'])
june_temps_df.set_index(june_temps_df['Date'], inplace=True)
june_temps_df = june_temps_df.sort_index()
print(june_temps_df.to_string(index=False))
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temps_df.describe()
```

```python
june_temps_df.plot.hist(bins=20)
plt.tight_layout()
```

```python
# Add rain data for further analysis
year = 2010
AllTheJunes = []
while (year <= 2017):
    junes = session.query(Measurement.date, Measurement.tobs, Measurement.prcp).filter(Measurement.date >= f'{year}-06-01' ).filter(Measurement.date <= f'{year}-06-30').all()
    AllTheJunes.append(junes)
    year += 1
june_temps = sum(AllTheJunes,[])
june_temps_df = pd.DataFrame(june_temps, columns=['Date', 'June Temp (degF)', 'June Precipitation (in)'])
june_temps_df.set_index(june_temps_df['Date'], inplace=True)
print(june_temps_df.to_string(index=False))
```

```python
# print out the summary statistics for the June temperature and rain data.
june_temps_df.describe()
```

```python
june_temps_df.plot(x='Date', y='Precipitation (in)', kind='line')
plt.xticks(rotation='vertical')
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
year = 2010
AllTheDecs = []
while (year <= 2017):
    decembers = session.query(Measurement.date, Measurement.tobs).filter(Measurement.date >= f'{year}-12-01' ).filter(Measurement.date <= f'{year}-12-31').all()
    AllTheDecs.append(decembers)
    year += 1
print(len(AllTheDecs))
```

```python
# 7. Convert the December temperatures to a list.
december_temps = sum(AllTheDecs,[])
print(december_temps)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_temps_df = pd.DataFrame(december_temps, columns=['Date', 'Dec Temp (degF)'])
dec_temps_df.set_index(dec_temps_df['Date'], inplace=True)
dec_temps_df = dec_temps_df.sort_index()
print(dec_temps_df.to_string(index=False))
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_temps_df.describe()
```

```python
dec_temps_df.plot.hist(bins=20)
plt.tight_layout()
```

```python
# Add rain data for December
year = 2010
AllTheDecs = []
while (year <= 2017):
    decs = session.query(Measurement.date, Measurement.tobs, Measurement.prcp).filter(Measurement.date >= f'{year}-12-01' ).filter(Measurement.date <= f'{year}-12-31').all()
    AllTheDecs.append(decs)
    year += 1
dec_temps = sum(AllTheDecs,[])
dec_temps_df2 = pd.DataFrame(dec_temps, columns=['Date', 'December Temp (degF)', ' Dec Precipitation (in)'])
dec_temps_df2.set_index(dec_temps_df['Date'], inplace=True)
print(dec_temps_df2.to_string(index=False))
```

```python
# print out the summary statistics for the Decemeber temperature and rain data
dec_temps_df2.describe()
```

```python
result = pd.concat([june_temps_df, dec_temps_df2])
result.plot(x="Date", y=['June Precipitation (in)', ' Dec Precipitation (in)'], kind='line')
plt.locator_params(axis='x', nbins=len(result['Date'])/400)
```
